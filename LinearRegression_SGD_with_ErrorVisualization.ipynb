```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([6, 8, 12, 14, 18])
y = np.array([350, 775, 1150, 1395, 1675])
# Assume the weight
W = np.array([1, 2])
```

```python
alpha = 0.001
h = [] # for predicted value
error = [] # for error value
iteration = 3
for j in range(iteration):
  for i in range(x.shape[0]):
    h.append(W[0] + W[1]*x[i])

    # data for visualizing error
    error.append((h[i] - y[i]) ** 2)

    # parameter update
    W[0] = W[0] - alpha * 2 * (h[i] - y[i])
    W[1] = W[1] - alpha * 2 * x[i] * (h[i] - y[i])
  plt.plot(x, h)
  h.clear()

# print(W)
plt.scatter(x, y)
plt.xlabel("Pizza Size (inch)")
plt.ylabel("Price (TK)")
plt.title("Linear Regression using Stochastic Gradient Descent")
plt.grid()
plt.show()
# visualization of error curve
plt.plot([i for i in range(iteration * x.shape[0])], error)
plt.xlabel("Epoch")
plt.ylabel("Squared Error")
plt.title("Error Curve Visualization")
plt.grid()
plt.show()
```
